Import 2 dataset, yaitu dataset test dan train 

```python
import pandas as pd

df_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Housing Price Model/test.csv')
df_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Housing Price Model/train.csv')
```

memilih kolom yang akan dijadikan sebagai feature

```python
df_train = df_train[['LotArea', 'OverallQual', 'OverallCond', 'YearBuilt','YearRemodAdd', 'TotalBsmtSF', 'GrLivArea', 'TotRmsAbvGrd', 'BedroomAbvGr', 'FullBath'
           , 'BsmtFinSF1', 'Fireplaces', 'GarageArea', 'GarageCars', 'PoolArea', 'MiscVal', 'YrSold'
           , 'Neighborhood', 'SaleType', 'SaleCondition','SalePrice']] 
```

```python
df_train.shape
```

memilih kolom yang akan dijadikan feature pada dataset test

```python
df_test = df_test[['LotArea', 'OverallQual', 'OverallCond', 'YearBuilt','YearRemodAdd', 'TotalBsmtSF', 'GrLivArea', 'TotRmsAbvGrd', 'BedroomAbvGr', 'FullBath'
           , 'BsmtFinSF1', 'Fireplaces', 'GarageArea', 'GarageCars', 'PoolArea', 'MiscVal', 'YrSold'
           , 'Neighborhood', 'SaleType', 'SaleCondition']] 
```

```python
df_test.shape
```

menangani missing value dengan agregasi mean

```python
#mengisi missing value di data test
df_test = df_test.fillna(value = {'TotalBsmtSF': df_test['TotalBsmtSF'].mean(), 'BsmtFinSF1': df_test['BsmtFinSF1'].mean(),
                         'GarageArea': df_test['GarageArea'].mean(), 
                         'GarageCars': df_test['GarageCars'].mean(),
                         'SaleType' :'WD'})
```

melakukan encoding untuk merubah data katagorikal menjadi data numerik

```python
#melakukan encoding untuk data katagorikal

import numpy as np
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
df_train['Neighborhood'] = LE.fit_transform(df_train['Neighborhood'])
print(LE.classes_)
print(np.sort(df_train['Neighborhood'].unique()))
print('==================================================================')

LE = LabelEncoder()
df_train['SaleType'] = LE.fit_transform(df_train['SaleType'])
print(LE.classes_)
print(np.sort(df_train['SaleType'].unique()))
print('==================================================================')

LE = LabelEncoder()
df_train['SaleCondition'] = LE.fit_transform(df_train['SaleCondition'])
print(LE.classes_)
print(np.sort(df_train['SaleCondition'].unique()))
print('==================================================================')
```

```python
#melakukan encoding untuk data katagorikal

import numpy as np
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
df_test['Neighborhood'] = LE.fit_transform(df_test['Neighborhood'])
print(LE.classes_)
print(np.sort(df_test['Neighborhood'].unique()))
print('==================================================================')

LE = LabelEncoder()
df_test['SaleType'] = LE.fit_transform(df_test['SaleType'])
print(LE.classes_)
print(np.sort(df_test['SaleType'].unique()))
print('==================================================================')

LE = LabelEncoder()
df_test['SaleCondition'] = LE.fit_transform(df_test['SaleCondition'])
print(LE.classes_)
print(np.sort(df_test['SaleCondition'].unique()))
print('==================================================================')
```

```python
df_train.corr().style.background_gradient().set_precision(2)
```

```python
X = df_train.drop(['SalePrice'], axis = 1)
y = df_train['SalePrice']
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state = 0)
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators = 10000, learning_rate = 0.095)
model.fit(X_train, y_train,
          early_stopping_rounds = 55, 
          eval_set = [(X_valid, y_valid)], 
          verbose = False)
```

```python
from sklearn.metrics import mean_absolute_error

val_predictions = model.predict(X_valid)
val_mae = mean_absolute_error(val_predictions, y_valid)
print('MAE adalah : ',val_mae)
```

```python
model.fit( df_train.drop(['SalePrice'], axis = 1), df_train['SalePrice'],
          early_stopping_rounds = 55, 
          eval_set = [(X_valid, y_valid)], 
          verbose = False)

test_preds = model.predict(df_test)
```

```python
output = pd.DataFrame({'Id': pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Housing Price Model/test.csv').Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```
